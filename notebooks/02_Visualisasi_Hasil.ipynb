# Visualisasi Hasil Eksperimen: RAG + FSM Validation
Notebook ini digunakan untuk menganalisis hasil dari `experiment_results.csv` dan menghasilkan grafik untuk laporan penelitian.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 1. Load Data
file_path = '../experiment_results.csv'
if not os.path.exists(file_path):
    print("File experiment_results.csv tidak ditemukan! Jalankan python main_experiment.py terlebih dahulu.")
else:
    df = pd.DataFrame(pd.read_csv(file_path))
    print(f"Data berhasil di-load: {len(df)} sampel")
```

## 1. Distribusi Status FSM
Melihat berapa banyak query yang berhasil lolos (`success`) dan berapa yang berhasil ditangkap sebagai error oleh FSM (`error`).

```python
plt.figure(figsize=(8, 5))
status_counts = df['fsm_status'].value_counts()
colors = ['#4CAF50', '#FF5252', '#FFC107']
status_counts.plot(kind='bar', color=colors)

plt.title('Distribusi Status Validasi FSM', fontsize=14)
plt.xlabel('Status', fontsize=12)
plt.ylabel('Jumlah Sampel', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 2. Analisis Efektivitas Validasi (Error Detection)
Menghitung persentase error yang berhasil dicegah oleh FSM.

```python
success_rate = (len(df[df['fsm_status'] == 'success']) / len(df)) * 100
error_rate = (len(df[df['fsm_status'] == 'error']) / len(df)) * 100

labels = ['Success (Lolos)', 'Error (Tertangkap FSM)']
sizes = [success_rate, error_rate]

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#4CAF50', '#FF5252'])
plt.title('Efektivitas Pipeline RAG + FSM', fontsize=14)
plt.show()
```

## 3. Contoh Kasus Tertangkap Error
Menampilkan contoh di mana Baseline RAG memberikan jawaban, namun FSM mendeteksi adanya ketidakvalidan (potensi halusinasi).

```python
error_samples = df[df['fsm_status'] == 'error'].head(3)
if not error_samples.empty:
    for i, row in error_samples.iterrows():
        print(f"Q: {row['question']}")
        print(f"Baseline Answer: {row['baseline_answer']}")
        print(f"FSM Status: {row['fsm_status']} (Error caught at: {row['fsm_states_passed'].split('|')[-1]})")
        print("-" * 50)
else:
    print("Tidak ada sampel error yang ditemukan.")
```
